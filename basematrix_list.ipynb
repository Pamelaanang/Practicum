# ***Create the relationship matrix***

```python
import pandas as pd

# Define the rows and columns as provided by the user
rows = columns = [
    "Data", "Database", "RDBMS", "Entity", "Field", "Record", "Relationship",
    "Primary_Key", "Foreign_Key", "ERD", "Dependent_Table", "Associative_Table",
    "Entity_Integrity", "Referential_Integrity", "1NF", "2NF", "SQL",
    "One-to-One_Relationship", "One-to-Many_Relationship", "Many-to-Many_Relationship",
    "Recursive_Relationship", "Supertype_Entity", "Subtype_Entity",
    "Disjoint_Rule", "Overlap_Rule", "Ternary_Relationship", "ACID"
]

# Create a 27x27 matrix with tuple pairings from rows and columns with the order switched
tuple_matrix = [[(columns[j], rows[i]) for j in range(len(columns))] for i in range(len(rows))]

# Convert the switched tuple matrix to a pandas DataFrame for better readability
tuple_matrix_df = pd.DataFrame(tuple_matrix, index=rows, columns=columns)

# Save the DataFrame to a CSV file
tuple_matrix_csv_path = 'tuple_matrix1.csv'
tuple_matrix_df.to_csv(tuple_matrix_csv_path)
```

```python
import pandas as pd

# Let's load the updated Excel file and extract the data from each sheet
file_path_updated = '/content/database.xlsx'

# Read each sheet into a separate dataframe
data1 = pd.read_excel(file_path_updated, sheet_name= 'Yan')
data2 = pd.read_excel(file_path_updated, sheet_name= 'Pamela')
data3 = pd.read_excel(file_path_updated, sheet_name= 'Yue')


```

```python
def dataframe_to_list(df):
    # Convert the entire DataFrame to string type, replace NaN with a placeholder
    return df.fillna('missing').astype(str).values.flatten().tolist()

matrix1_text = dataframe_to_list(data1)
matrix2_text = dataframe_to_list(data2)
matrix3_text = dataframe_to_list(data3)
```

# ***Clean all the extra " ", and also fill the blanks with NA.***

```python
matrices = [matrix1_text, matrix2_text, matrix3_text]

for i, matrix in enumerate(matrices, start=1):
    cleaned_matrix = [item.replace('"', '').replace(',', '').strip() if item.strip('"').strip() != '' else 'NAa' for item in matrix]
    cleaned_matrix = ['NA' if item == '' else item for item in cleaned_matrix]
    print(f"Cleaned matrix_{i}: {cleaned_matrix}")


```
